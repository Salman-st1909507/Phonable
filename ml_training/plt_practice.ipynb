```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
mpl.rcParams['lines.markersize']=5
counter=0
```

```python
def plot(x,y,
         title='',xlabel='',ylabel='',
         color='b', linestyle='-', linewidth=1,
         marker='.', markersize=5, label=''):
    
    plt.plot(x, y, 
             color=color,
             linestyle=linestyle,
             marker=marker,
             markersize=markersize,
             linewidth=linewidth,
             label=label)
    
    # plt.plot(x, y,label=label)
    
    plt.xticks(np.linspace(0,20,21))
    plt.yticks(np.linspace(1,100,10))
    
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

    plt.legend()
    plt.grid(True)

styles=plt.style.available
print(styles)
plt.style.use(styles[counter])
print(f'counter: {counter}')
counter=(counter+1)%len(styles)

x= np.linspace(1,20,16)
x.sort()

y1=np.random.randint(1,100,16)
y1.sort()

fig=plt.figure(figsize=(15,50), facecolor='#dddddd')
# ax=plt.subplot(3,2,1)
# plot(x, y1,'Plot','X','Y', color='b', linestyle='-', marker='v',linewidth=2, label='Random Data 1')

y2=np.random.randint(1,100,16)
y2.sort()

# plt.subplot(3,2,3)
# plot(x, y2,'Plot','X','Y', color='g', linestyle='--', marker='8',linewidth=3, label='Random Data 2')


y3=np.random.randint(1,100,16)
y3.sort()

# plt.subplot(2,2,2)
# plot(x, y3,'Plot','X','Y', color='y', linestyle=':', marker='h',linewidth=1, label='Random Data 3')

y4=np.random.randint(1,100,16)
y4.sort()

# plt.subplot(3,1,3)
# plot(x, y4,'Plot','X','Y', color='r', linestyle='-.', marker='1',linewidth=1, label='Random Data 4')

for index in range(len(styles)):
    plt.style.use(styles[index])

    plt.subplot(15,2,index+1)
    plot(x, y1,f'{styles[index]} {index}','X','Y', color='b', linestyle='-', marker='v',linewidth=2, label='Random Data 1')
    plot(x, y2,f'{styles[index]} {index}','X','Y', color='g', linestyle='--', marker='8',linewidth=3, label='Random Data 2')
    plot(x, y3,f'{styles[index]} {index}','X','Y', color='y', linestyle=':', marker='h',linewidth=1, label='Random Data 3')
    plot(x, y4,f'{styles[index]} {index}','X','Y', color='r', linestyle='-.', marker='1',linewidth=1, label='Random Data 4')


plt.tight_layout()
```
